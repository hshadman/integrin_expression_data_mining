```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sys
from scipy.stats import pearsonr
from scipy.stats import spearmanr

#This cell just reads in and prepares data
#The output dataframes are ccle_rna_data and protein_data 

ccle_rna_data = pd.read_csv('CCLE_expression.csv',sep=",")
ccle_p_data = pd.read_csv('proteome/protein_quant_current_normalized.csv',sep=",")

data_header = list(ccle_p_data.columns)

#I had to make some modifications here to make this data file agree with the data on depmap portal
drop_columns = ['TenPx01_Peptides', 'TenPx02_Peptides', 'TenPx03_Peptides', 'TenPx04_Peptides', 'TenPx06_Peptides', 'TenPx07_Peptides', 'TenPx08_Peptides', 'TenPx09_Peptides', 'TenPx10_Peptides', 'TenPx11_Peptides', 'TenPx12_Peptides', 'TenPx13_Peptides', 'TenPx16_Peptides', 'TenPx17_Peptides', 'TenPx19_Peptides', 'TenPx20_Peptides', 'TenPx21_Peptides', 'TenPx22_Peptides', 'TenPx27_Peptides', 'TenPx28_Peptides', 'TenPx29_Peptides', 'TenPx33_Peptides', 'TenPx34_Peptides', 'TenPx35_Peptides', 'TenPx36_Peptides', 'TenPx37_Peptides', 'TenPx38_Peptides', 'TenPx39_Peptides', 'TenPx40_Peptides', 'TenPx42_Peptides', 'TenPx23_Peptides', 'TenPx05_Peptides', 'TenPx30_Peptides', 'TenPx31_Peptides', 'TenPx32_Peptides', 'TenPx14_Peptides', 'TenPx15_Peptides', 'TenPx41_Peptides', 'TenPx26_Peptides', 'TenPx25_Peptides', 'TenPx18_Peptides', 'TenPx24_Peptides']
all_data = ccle_p_data.drop(drop_columns,axis=1)

temp_a = all_data["CAL120_BREAST_TenPx02"]
temp_b = all_data["CAL120_BREAST_TenPx28"]
temp = []
for i in range(len(temp_a)):
    if np.isnan(temp_a.iloc[i]) and np.isnan(temp_b.iloc[i]):
        temp.append(np.nan)
    elif np.isnan(temp_a.iloc[i]):
        temp.append(temp_b.iloc[i])
    elif np.isnan(temp_b.iloc[i]):
        temp.append(temp_a.iloc[i])
    else:
        temp.append((temp_a.iloc[i] + temp_b.iloc[i])/2)
all_data = all_data.drop(["CAL120_BREAST_TenPx02","CAL120_BREAST_TenPx28"],axis=1)
all_data["CAL120_BREAST_TenPx0228"] = temp

temp_a = all_data["HCT15_LARGE_INTESTINE_TenPx30"]
temp_b = all_data["HCT15_LARGE_INTESTINE_TenPx18"]
temp = []
for i in range(len(temp_a)):
    if np.isnan(temp_a.iloc[i]) and np.isnan(temp_b.iloc[i]):
        temp.append(np.nan)
    elif np.isnan(temp_a.iloc[i]):
        temp.append(temp_b.iloc[i])
    elif np.isnan(temp_b.iloc[i]):
        temp.append(temp_a.iloc[i])
    else:
        temp.append((temp_a.iloc[i] + temp_b.iloc[i])/2)
all_data = all_data.drop(["HCT15_LARGE_INTESTINE_TenPx30","HCT15_LARGE_INTESTINE_TenPx18"],axis=1)
all_data["HCT15_LARGE_INTESTINE_TenPx3018"] = temp

temp_a = all_data["SW948_LARGE_INTESTINE_TenPx11"]
temp_b = all_data["SW948_LARGE_INTESTINE_TenPx20"]
temp = []
for i in range(len(temp_a)):
    if np.isnan(temp_a.iloc[i]) and np.isnan(temp_b.iloc[i]):
        temp.append(np.nan)
    elif np.isnan(temp_a.iloc[i]):
        temp.append(temp_b.iloc[i])
    elif np.isnan(temp_b.iloc[i]):
        temp.append(temp_a.iloc[i])
    else:
        temp.append((temp_a.iloc[i] + temp_b.iloc[i])/2)
all_data = all_data.drop(["SW948_LARGE_INTESTINE_TenPx11","SW948_LARGE_INTESTINE_TenPx20"],axis=1)
all_data["SW948_LARGE_INTESTINE_TenPx1120"] = temp

protein_data = all_data.copy()
#ccle_rna_data_backup = ccle_rna_data.copy()


```

```python
#This cell extracts and sorts the data so that the protein and RNA data have data from the same cell
#  types in the same order

CCLE_sample_info = pd.read_csv('sample_info.csv')
drop_cols= ['cell_line_name', 'stripped_cell_line_name',
       'alias', 'COSMICID', 'sex', 'source', 'Achilles_n_replicates',
       'cell_line_NNMD', 'culture_type', 'culture_medium', 'cas9_activity',
       'RRID', 'WTSI_Master_Cell_ID', 'sample_collection_site',
       'primary_or_metastasis', 'primary_disease', 'Subtype', 'age',
       'Sanger_Model_ID', 'depmap_public_comments', 'lineage',
       'lineage_subtype', 'lineage_sub_subtype', 'lineage_molecular_subtype']
CCLE_sample_info = CCLE_sample_info.drop(drop_cols,axis=1)

#protein_data = protein_data.drop(['Protein_Id','Gene_Symbol','Description','Group_ID','Uniprot','Uniprot_Acc'],axis=1)
protein_cols = protein_data.columns.to_list()
protein_cols = protein_cols[6:]

#There are a couple of cells in the protein data that are missing from the RNA data
#I will remove those here
#ACH-001039,ACH-000979,ACH-000309,ACH-001198,ACH-000727,ACH-001171
cols_to_drop = ['ACH-001039','ACH-000979','ACH-000309','ACH-001198','ACH-000727','ACH-001171']

protein_depmap_ids = []
ccle_ids_to_drop = []
for i in range(len(protein_cols)):
    temp = protein_cols[i].split('_')
    temp = '_'.join(temp[0:-1])
    temp = CCLE_sample_info.loc[CCLE_sample_info["CCLE_Name"] == temp]["DepMap_ID"].values
    if temp[0] not in cols_to_drop: 
        protein_depmap_ids.append(temp)
    else:
        ccle_ids_to_drop.append(protein_cols[i])
protein_data = protein_data.drop(ccle_ids_to_drop,axis=1)

ccle_rna_selected = pd.DataFrame(columns = ccle_rna_data.columns)
for i in range(len(protein_depmap_ids)):
    temp = ccle_rna_data.loc[ccle_rna_data["Unnamed: 0"] == protein_depmap_ids[i][0]]
    ccle_rna_selected = ccle_rna_selected.append(temp,ignore_index=True)

#protein_data_original = protein_data.copy()
#ccle_rna_selected_original = ccle_rna_selected.copy()
protein_data = protein_data.T
```

```python
#Now ccle_rna_selected has RNA data and protein_data has protein data for the same cells in the same order
#  but the genes in the frames are not the same and are not in order
#So, I need to do the same for the genes-- extract and sort to get the same genes in the same order
#   for both RNA and protein data

geneid_uniprot_map = pd.read_csv('unmatched_uniprot_geneid_map.csv',sep=",")
map_gene_id = geneid_uniprot_map["Gene ID"].to_list()
map_gene_symbols = geneid_uniprot_map["Gene_Symbol"].to_list()
map_uniprot = geneid_uniprot_map["Uniprot_Acc"].to_list()

protein_data.columns = protein_data.iloc[1]
p_columns = protein_data.columns
p_index = protein_data.index
ccle_rna_selected.index = ccle_rna_selected.iloc[:,0]
rna_columns = ccle_rna_selected.columns
rna_index = ccle_rna_selected.index

protein_uniprot = protein_data.loc["Uniprot_Acc"]

protein_genes = p_columns.to_list()
rna_genes = rna_columns.to_list()
for i in range(1,len(rna_genes)):
    temp = rna_genes[i].split(" ")
    temp2 = int(temp[1][1:-1])
    for j in range(len(map_gene_id)):
        if temp2 == map_gene_id[j]:
            temp[0] = str(map_gene_symbols[j])
            rna_genes[i] = ' '.join(temp)
    
rna_gene_names = []
for i in range(len(rna_genes)):
    temp = rna_genes[i].split(" ")
    #Making some modifications here so things agree with DepMap data.
    if temp[0] == 'ATP5F1B':
        temp[0] = 'ATP5B'
    if temp[0] == 'ECPAS':
        temp[0] = 'ECM29'
    if temp[0] == 'EPDR1':
        temp[0] = 'UCC1'
    if temp[0] == 'GMCL2':
        temp[0] = 'GMCL1P1'
    if temp[0] == 'HDGFL2':
        temp[0] = 'HDGFRP2'
    if temp[0] == 'MIA2':
        temp[0] = 'CTAGE5'
    if temp[0] == 'PATJ':
        temp[0] = 'INADL'
    if temp[0] == 'PTPA':
        temp[0] = 'PPP2R4'
    if temp[0] == 'ZNF724':
        temp[0] = 'ZNF724P'
    if len(temp) > 2:
        print(temp)
    else:
        rna_gene_names.append(temp[0])
    


#Here, I get the rows for each gene in both data sets.
#  In some cases, there are multiple proteins for a gene. I keep all of the proteins here.
#  So, a single RNA may be on multiple rows, one for each of the proteins.
# This is consistent with the data in DepMap.
# I also get a list of protein gene names which do not match any RNA gene names
#  I do not do anything with that here
gene_locations = []
no_matches = []
no_match_index = []
for i in range(len(protein_genes)):
    test = 0
    for j in range(len(rna_gene_names)):
        if protein_genes[i] == rna_gene_names[j]:
            gene_locations.append([protein_genes[i],i,j])
            test = 1
    if test == 0:
        no_matches.append(protein_genes[i])
        no_match_index.append(i)
        
protein_data_unmatched = protein_data.iloc[:,no_match_index].copy()
protein_data_unmatched.loc["Uniprot_Acc"].to_csv("unmatched_uniprot_new.csv")      
        
```

```python
#This cell makes new data frames so that the cells are in order
#It takes ~10 minutes to run on my computer for some reason
#ccle_rna_selected = ccle_rna_selected.drop(["Unnamed: 0"],axis = 1)
protein_data_matched = pd.DataFrame(columns=p_index)
rna_data_matched = pd.DataFrame(columns=rna_index)
#print(len(gene_locations))
for i in range(len(gene_locations)):
    #print(i)
    protein_data_matched.loc[i] = protein_data.iloc[:,gene_locations[i][1]]
    rna_data_matched.loc[i] = ccle_rna_selected.iloc[:,gene_locations[i][2]]
    
#protein_data_matched_backup = protein_data_matched.copy()
#rna_data_matched_bakceup = rna_data_matched.copy()
```

```python
#Now calculate and output the actual correlations

output = []
for i in range(protein_data_matched.shape[0]):
    temp = []
    p_i = protein_data_matched.iloc[i].copy()
    r_i = rna_data_matched.iloc[i].copy()
    gene_symbol = p_i["Gene_Symbol"]
    uniprot_acc = p_i["Uniprot_Acc"]
    temp.append(gene_symbol)
    temp.append(uniprot_acc)
    #if p_i["Uniprot_Acc"] == 'P49588':
    #    print(p_i)
    #    print(r_i)
    #r_i = r_i.drop(["Unnamed: 0"])
    p_i = p_i.drop(["Protein_Id","Gene_Symbol","Description","Group_ID","Uniprot","Uniprot_Acc"])
    p_i = p_i.to_list()
    r_i = r_i.to_list()
    pair = pd.DataFrame(list(zip(p_i,r_i)))
    pair_drop = pair.dropna()
    #if gene_symbol == 'HOXD13':
    #    pd.DataFrame(pair_drop).to_csv("HOXD13_data.csv")
    #    sys.exit()
    if len(pair_drop) < 5:
        print(gene_symbol)
        print(pair_drop)
    else:
        #print(gene_symbol)
        #print(pair_drop)
        pear_r,pear_p = pearsonr(pair_drop.iloc[:,0],pair_drop.iloc[:,1])
        spear_r, spear_p = spearmanr(pair_drop.iloc[:,0],pair_drop.iloc[:,1])
        temp.append(pear_r)
        temp.append(pear_p)
        temp.append(spear_r)
        temp.append(spear_p)
        temp.append(len(pair_drop))
        output.append(temp)
    

out_columns = ["GeneName","Uniprot_Acc","PearsonR","PearsonP","SpearmanR","SpearmanP","NumberCells"]
pd.DataFrame(output,columns=out_columns).to_csv("protein_rna_correlation.csv")
```
